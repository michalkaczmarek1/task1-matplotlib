```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

movies = pd.read_csv('tmdb_movies.csv')
genres = pd.read_csv('tmdb_genres.csv')
movies.drop('Unnamed: 0', axis=1, inplace=True)
genres.rename(columns={'Unnamed: 0':'id'}, inplace=True)
genres.dropna(inplace=True)
genres
```

<b>Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 (włącznie) do 2016 roku (włącznie). Następnie na tej podstawie stwórz wykres, w którym średnie przychody są wykresem kolumnowym, a średni budżet wykresem liniowym na tych samych osiach. Sformatuj odpowiednio oś X oraz oś Y. Dodaj tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi.<b>

```python
movies['year'] = pd.to_datetime(movies['release_date'])
movies['year'] = movies['year'].dt.year
movies['year'] = movies['year'].fillna(value=0)
movies['year'] = movies['year'].astype(int)
movies_with_specific_period = movies[(movies['year'] >= 2010) & (movies['year'] <= 2016)].pivot_table(values=['budget', 'revenue'], columns='year').transpose()
movies_with_specific_period
```

```python
def prepare_data_budget_and_revenue(x):
    x = round(float(x)/1000000, 2)
    return x
```

```python
movies_with_specific_period['budget'] = movies_with_specific_period['budget'].apply(lambda x: prepare_data_budget_and_revenue(x))
movies_with_specific_period['revenue'] = movies_with_specific_period['revenue'].apply(lambda x: prepare_data_budget_and_revenue(x))
```

```python
movies_with_specific_period
```

```python
def million(x, pos):
    return '{:2.1f}M'.format(x)
```

```python
fig, ax = plt.subplots()
formatter = plt.FuncFormatter(million)
ax.yaxis.set_major_formatter(million)
ax.bar(movies_with_specific_period.index,movies_with_specific_period['revenue'], label='revenues')
ax.plot(movies_with_specific_period.index,movies_with_specific_period['budget'], color='red', lw=3, label='budget')
ax.set_title("Średni przychód i budżet filmu w latach 2010-2016 roku")
ax.legend(loc=(1.05,0.80))
```

<b>Baza filmów zawiera kolumnę z id gatunku (genre_id). Na tej podstawie połącz ze sobą bazę filmów z bazą gatunków, tak aby w bazie filmów można było odczytać nazwę gatunku filmu.<b>

```python
genres.rename(columns={'id':'genre_id'}, inplace=True)
movies = pd.merge(movies, genres, how='inner', on='genre_id')
movies
```

<b>Jaki gatunek filmu z bazy pojawia się w niej najczęściej? Ile filmów tego gatunku znajduje się w bazie?</b>

```python
movies_genres_collection = movies.pivot_table(values='genre_id', columns='genres', aggfunc=len).transpose()
movies_genres_collection.rename(columns={'genre_id':'amount_movies'}, inplace=True)
movies_genres_collection.sort_values(by='amount_movies', ascending=False)
```

Najczęściej pojawia się gatunek "Drama". Ilość filmów tego gatunku w bazie to 1207.

<b>Filmy, którego gatunku trwają średnio najdłużej (runtime)?</b>

```python
movies_genres_runtime_collection = movies.pivot_table(values='runtime', columns='genres').transpose()
movies_genres_runtime_collection.sort_values(by='runtime', ascending=False)
```

Filmy z gatunku "History" trwają średnio najdłużej.
